# FEATURE ENGINEERING

```python
# Lets combine the Date_time column and Time_hr column together

data['DateTime'] = pd.to_datetime(data['Date_Time'].astype(str) + ' ' + data['Time_Hr'].astype(str))
```

```python
data.head()
```

```python
# Lets drop the  "Date_Time" and "Time_Hr" columns because we no longer need them
data.drop(columns=['Date_Time'], inplace=True)
data.drop(columns=['Time_Hr'], inplace=True)
```

```python
data = data[['DateTime'] + [col for col in data.columns if col not in ['DateTime']]]
```

```python
# Lets Inspect the Date_Time column to see if those two columns were combine successfully

print(data['DateTime'].unique())
data.info()
```

```python
# Extract hour, day_of_week, and month
data['hour'] = data['DateTime'].dt.hour
data['day_of_week'] = data['DateTime'].dt.dayofweek  # 0=Monday, 6=Sunday
data['month'] = data['DateTime'].dt.month  # 1=January, 12=December
```

```python


import numpy as np

data['hour_sin'] = np.sin(2 * np.pi * data['hour'] / 24)
data['hour_cos'] = np.cos(2 * np.pi * data['hour'] / 24)
data['day_of_week_sin'] = np.sin(2 * np.pi * data['day_of_week'] / 7)
data['day_of_week_cos'] = np.cos(2 * np.pi * data['day_of_week'] / 7)
```

```python
data.head(10)
```

# TRAIN TEST SPLIT

```python
from sklearn.model_selection import train_test_split, GridSearchCV
# Train-Test Split
X = data.drop(columns=['Residual Demand', 'DateTime', 'hour', 'day_of_week' ])  # Features
y = data['Residual Demand']  # Target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

# MODEL SELECTION

# MODEL TRAINING AND EVALUATION

## RandomForestRegressor & XGBRegressor

```python
from sklearn.ensemble import RandomForestRegressor
# Train Random Forest Regressor
rf = RandomForestRegressor(n_estimators=100, max_depth=10, random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
```

```python
from xgboost import XGBRegressor
# Train XGBoost Regressor
xgb = XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=10, random_state=42)
xgb.fit(X_train, y_train)
y_pred_xgb = xgb.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Evaluate models
def evaluate(y_true, y_pred, model_name):
    print(f"{model_name} Performance")
    print("MAE:", mean_absolute_error(y_true, y_pred))
    print("MSE:", mean_squared_error(y_true, y_pred))
    print("R^2:", r2_score(y_true, y_pred))
    print("\n")

evaluate(y_test, y_pred_rf, "Random Forest")
evaluate(y_test, y_pred_xgb, "XGBoost")
```
